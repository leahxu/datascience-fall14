```python
players = pd.read_csv('players.csv')
countries = pd.read_csv('countries.csv')
events = pd.read_csv('events.csv')

df = pd.merge(countries, players, on='country_id')

df.rename(columns={'name_x':'country_name'}, inplace=True)

df['num_players'] = df.groupby('country_id')['country_id'].transform('count')

df[['country_name', 'num_players']].drop_duplicates()
```

```python
player1975 = players[pd.to_datetime(players.birthdate).apply(lambda x:x.year) == 1975][['country_id', 'player_id']]
players_count = pd.merge(countries, player1975, on='country_id', how='outer').groupby('name').count()['player_id']

df = pd.DataFrame(players_count)
df.reset_index(level=0, inplace=True)
df.rename(columns={'player_id':'num_player'}, inplace=True)
df.rename(columns={'name':'country_name'}, inplace=True)
df
```
