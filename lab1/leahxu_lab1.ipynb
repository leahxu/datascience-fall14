# Exercise 3: Linear regression on the boston house prices dataset

## Part I

```python
# Import linear regression and the diabetes dataset
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt

# Load the diabetes dataset
boston = load_boston()

# (a) 20% training, 80% testing, (b) 40% training, 60% testing, (c) 60% training, 40% testing, and (d) 80% training, 20% testing
percent_training = [.2, .4, .6, .8]

residual_sum = []

for i in range(0, len(percent_training)): 
    
    # Split the data into training/testing sets
    boston_X_train = boston.data[:int(percent_training[i]*len(boston.data))]
    boston_X_test = boston.data[int(percent_training[i]*len(boston.data)):]
    
    # Split the targets into training/testing sets
    boston_y_train = boston.target[:int(percent_training[i]*len(boston.data))]
    boston_y_test = boston.target[int(percent_training[i]*len(boston.data)):]

    # Create linear regression object
    regr = LinearRegression()

    # Train the model using the training sets
    regr.fit(boston_X_train, boston_y_train)
    
    temp_residual_sum = np.mean((regr.predict(boston_X_test) - boston_y_test) ** 2)
    residual_sum.append(temp_residual_sum)
    
    print "Residual sum of Squares", temp_residual_sum, "for", percent_training[i], "training"
    
plt.scatter(percent_training, residual_sum)
plt.title("Residual Sum of Squares Over Percentage of Training Data")
plt.xlabel("Percentage of Training Data")
plt.ylabel("Residual Sum of Squares")
plt.grid(True)

print
print "The optimal percentage of training data is 80 percent with 20 percent testing."
```

# Exercise 5: Clustering the iris data

```python
# Loading the iris data
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn import metrics

iris = load_iris()

X = iris.data[:]
y = iris.target[:]

kmeans = KMeans(init='random', n_clusters = 2, n_init=10)
kmeans.fit(X)

# Homogeneity score: evaluates if each cluster contains only memebers of a single class
print "Homogeneity_score = ", metrics.homogeneity_score(y, kmeans.labels_)

# Completeness score: evaluates if all members of a given class are assigned to the same cluster
print "Completeness_score = ", metrics.completeness_score(y, kmeans.labels_)
```
