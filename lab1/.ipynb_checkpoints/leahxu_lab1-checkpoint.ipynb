```python
# Exercise 2.2:

# Import recursive feature elimination, linear regression and the diabetes dataset
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()
residual_sum = []
num_features = list(xrange(1, 11))

for i in range(1,11):
    # Split the data into training/testing sets
    diabetes_X_train = diabetes.data[:-20]
    diabetes_X_test = diabetes.data[-20:]

    # Split the targets into training/testing sets
    diabetes_y_train = diabetes.target[:-20]
    diabetes_y_test = diabetes.target[-20:]
    
    # Create the RFE object using i features and rank the features for the diabetes datasets
    rfe = RFE(estimator=regr, n_features_to_select=i, step=1)
    rfe.fit(diabetes_X_train,diabetes_y_train)
    residual_sum.append(np.mean((rfe.predict(diabetes_X_test) - diabetes_y_test) ** 2))

    
plt.scatter(num_features, residual_sum)
plt.title("Residual Sum of Squares Over Features")
plt.xlabel("Number of Features")
plt.ylabel("Residual Sum of Squares")
plt.grid(True)

print "The optimal number of features to use is 3."
```

```python
# Exercise 3: Linear regression on the boston house prices dataset
# Exercise 3.1

# Import linear regression and the diabetes dataset
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt

# Load the diabetes dataset
boston = load_boston()

# (a) 20% training, 80% testing, (b) 40% training, 60% testing, (c) 60% training, 40% testing, and (d) 80% training, 20% testing
percent_training = [.2, .4, .6, .8]

residual_sum = []

for i in range(0, len(percent_training)): 
    
    # Split the data into training/testing sets
    boston_X_train = boston.data[:int(percent_training[i]*len(boston.data))]
    boston_X_test = boston.data[int(percent_training[i]*len(boston.data)):]
    
    # Split the targets into training/testing sets
    boston_y_train = boston.target[:int(percent_training[i]*len(boston.data))]
    boston_y_test = boston.target[int(percent_training[i]*len(boston.data)):]

    # Create linear regression object
    regr = LinearRegression()

    # Train the model using the training sets
    regr.fit(boston_X_train, boston_y_train)

    residual_sum.append(np.mean((regr.predict(boston_X_test) - boston_y_test) ** 2))
    
plt.scatter(percent_training, residual_sum)
plt.title("Residual Sum of Squares Over Percentage of Training Data")
plt.xlabel("Percentage of Training Data")
plt.ylabel("Residual Sum of Squares")
plt.grid(True)

print "The optimal percentage of training data is 80 percent with 20 percent testing."
```

```python
# Exercise 4: Classification on the Iris dataset

# Loading the iris data
from sklearn.datasets import load_iris
from sklearn import svm

print "Linear SVM"

iris = load_iris()

X = iris.data[:]
y = iris.target[:]

X_train = iris.data[:int(len(iris.data)*.8)]
y_train = iris.target[:int(len(iris.data)*.8)]

X_test = iris.data[int(len(iris.target)*.8):]
y_test = iris.target[int(len(iris.target)*.8):]

# Create linear SVM 
clf_linear = svm.SVC(kernel='linear')

# Fit linear SVM
clf_linear.fit(X_train, y_train)

# Predict class for testing data
pred = clf_linear.predict(X_test)

# Calculate prediction error for testing data
errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        errors += 1.0
print "Test data errors = ",errors, "out of", len(pred), "predictions"

# Calculate prediction error for all points
total_pred = clf_linear.predict(X)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != y[i]:
        errors += 1.0
print "Total errors = ",errors, "out of", len(total_pred), "predictions"
```

```python
# Create non-linear SVM using the Radial Basis Kernel function

print "Non-linear SVM using the Radial Basis Kernel function"

clf_non_linear = svm.SVC(kernel='rbf')

# Fit the svm

clf_non_linear.fit(X_train,y_train)

# Calculate prediction error for testing data
pred = clf_non_linear.predict(X_test)
errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        errors += 1.0
print "Test data errors = ",errors, "out of", len(pred), "predictions"

# Calculate prediction error for all points
total_pred = clf_non_linear.predict(X)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != y[i]:
        errors += 1.0
print "Total errors = ",errors, "out of", len(total_pred), "predictions"
```

```python
# Exercise 5: Clustering the iris data

# Loading the iris data
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np

iris = load_iris()

X = iris.data[:]
y = iris.target[:]

X_train = iris.data[:int(len(iris.data)*.8)]
y_train = iris.target[:int(len(iris.data)*.8)]

X_test = iris.data[int(len(iris.target)*.8):]
y_test = iris.target[int(len(iris.target)*.8):]

kmeans = KMeans(init='random', n_clusters = 2, n_init=10)
kmeans.fit(X)

# Homogeneity score: evaluates if each cluster contains only memebers of a single class
print "Homogeneity_score = ",metrics.homogeneity_score(y,kmeans.labels_)

# Completeness score: evaluates if all members of a given class are assigned to the same cluster
print "Completeness_score = ",metrics.completeness_score(y,kmeans.labels_)
```
